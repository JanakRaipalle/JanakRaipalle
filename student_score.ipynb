```python
# Step 1: Import Libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Step 2: Define Study Hours (X) and Test Scores (Y)
X = np.array([2, 4, 6, 8, 10]).reshape(-1, 1)  # Study Hours (input)
Y = np.array([50, 80, 110, 140, 170])  # Test Scores (output)

# Step 3: Train the Linear Regression Model
model = LinearRegression()
model.fit(X, Y)

# Step 4: Make Predictions
Y_pred = model.predict(X)

# Step 5: Plot the Results
plt.scatter(X, Y, color='blue', label='Actual Scores')  # Actual data points
plt.plot(X, Y_pred, color='red', linewidth=2, label='Predicted Line')  # Regression line
plt.xlabel('Study Hours')
plt.ylabel('Test Score')
plt.title('Linear Regression: Study Hours vs Test Score')
plt.legend()
plt.show()

# Step 6: Predict a New Score
new_hours = np.array([[7]])  # Example: If a student studies 7 hours
predicted_score = model.predict(new_hours)
print(f"Predicted Score for 7 hours of study: {predicted_score[0]:.2f}")
```

```python
from fastapi import FastAPI, HTTPException
from fastapi.middleware.cors import CORSMiddleware
import sqlite3
import pandas as pd
from surprise import Dataset, Reader, SVD
from surprise.model_selection import train_test_split
from surprise import accuracy
```

```python
# Initialize FastAPI app
app = FastAPI()

# Enable CORS
app.add_middleware(
    CORSMiddleware,
    allow_origins=["*"],
    allow_credentials=True,
    allow_methods=["*"],
    allow_headers=["*"],
)

# Function to connect to the database
def get_db_connection():
    conn = sqlite3.connect("movies.db")
    conn.row_factory = sqlite3.Row
    return conn

# Load sample ratings dataset
ratings_data = [
    (1, "Sholay", 5), (1, "3 Idiots", 4), (1, "Queen", 3),
    (2, "Zindagi Na Milegi Dobara", 5), (2, "Gully Boy", 4),
    (3, "Kabir Singh", 5), (3, "Dilwale Dulhania Le Jayenge", 4),
]
ratings_df = pd.DataFrame(ratings_data, columns=["user_id", "movie", "rating"])

# Convert dataset into Surprise format
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(ratings_df[['user_id', 'movie', 'rating']], reader)

# Train SVD model
trainset, testset = train_test_split(data, test_size=0.2)
model = SVD()
model.fit(trainset)

# API Endpoint: Get all movies
@app.get("/movies")
def get_movies():
    conn = get_db_connection()
    cursor = conn.cursor()
    cursor.execute("SELECT * FROM movies")
    movies = cursor.fetchall()
    conn.close()

    if not movies:
        raise HTTPException(status_code=404, detail="No movies found")

    return [{"id": row["id"], "title": row["title"], "year": row["year"], "genre": row["genre"]} for row in movies]

# API Endpoint: Get recommendations for a user
@app.get("/recommend/{user_id}")
def get_recommendations(user_id: int):
    print(f"Received request for user {user_id}")
    all_movies = list(set(ratings_df["movie"]))
    user_movies = ratings_df[ratings_df["user_id"] == user_id]["movie"].tolist()

    predictions = []
    for movie in all_movies:
        if movie not in user_movies:
            pred = model.predict(uid=user_id, iid=movie)
            predictions.append((movie, pred.est))

    recommendations = sorted(predictions, key=lambda x: x[1], reverse=True)[:5]

    if not recommendations:
        raise HTTPException(status_code=404, detail="No recommendations found")

    return [{"title": movie, "predicted_rating": rating} for movie, rating in recommendations]

```

```python
# small project
from surprise import SVD
from surprise import Dataset, Reader
from surprise.model_selection import train_test_split
from surprise import accuracy

# Load dataset
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(ratings_df[['user_id', 'movie', 'rating']], reader)

# Split and train
trainset, testset = train_test_split(data, test_size=0.2)
model = SVD()
model.fit(trainset)

# Evaluate
predictions = model.test(testset)
accuracy.rmse(predictions)
```
